```python
import pandas as pd
import sqlite3
```

```python
db_file = '..\wildfire_data\FPA_FOD_20170508.sqlite'
conn = sqlite3.connect(db_file, isolation_level=None,
                           detect_types=sqlite3.PARSE_COLNAMES)
db_df = pd.read_sql("SELECT * FROM Fires", conn)
db_df_CA = db_df.loc[db_df['STATE']=='CA']
db_df_CA.to_csv('wildfire_data_CA.csv', index=False)
```
